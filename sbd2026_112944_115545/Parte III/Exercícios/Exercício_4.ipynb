```python
import pandas as pd
import sqlite3
```

```python
def atualizar_descricao_gene(gene_id, nova_descricao):
    conn = None
    try:
        conn = sqlite3.connect('disgenet.db')
        cursor = conn.cursor()
        
        # Executa a atualização utilizando prepared statements
        query = "UPDATE DIM_Gene SET geneDescription = ? WHERE geneId = ?"
        cursor.execute(query, (nova_descricao, gene_id))
        
        # Se nenhuma linha foi afetada, o ID não existia
        if cursor.rowcount == 0:
            print(f"Aviso: O Gene ID {gene_id} não foi encontrado na base de dados.")
        else:
            conn.commit()
            print(f"Sucesso: Descrição do Gene ID {gene_id} atualizada com êxito!")
            
    except sqlite3.DatabaseError as db_error:
        print(f"Erro de integridade ou base de dados: {db_error}")
        if conn:
            print("A reverter alterações (Rollback)...")
            conn.rollback()
    except Exception as general_error:
        print(f"Erro inesperado no sistema: {general_error}")
        if conn:
            conn.rollback()
    finally:
        if conn:
            conn.close()
            print("Ligação à base de dados terminada em segurança.")

if __name__ == "__main__":
    atualizar_descricao_gene("472", "Ataxia telangiectasia mutated protein kinase.")
```
