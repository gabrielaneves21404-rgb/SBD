```python
import pandas as pd
import sqlite3
```

```python


# Lista de ficheiros que carregou na Parte 1
csv_files = {
    "disease2class": "disease2class.csv",
    "diseaseAttributes": "diseaseAttributes.csv",
    "diseaseClass": "diseaseClass.csv",
    "geneAttributes": "geneAttributes.csv",
    "geneDiseaseNetwork": "geneDiseaseNetwork.csv",
    "variantAttributes": "variantAttributes.csv",
    "variantDiseaseNetwork": "variantDiseaseNetwork.csv"
}

conn = sqlite3.connect('disgenet.db')

print("A carregar os ficheiros CSV para a base de dados 'disgenet.db'...")
for table_name, csv_path in csv_files.items():
    df = pd.read_csv(csv_path)
    # Grava o DataFrame como uma tabela SQL
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    print(f"-> Tabela '{table_name}' criada com sucesso!")

conn.close()
print("Base de dados totalmente inicializada e pronta a usar!")
```

```python
def consultar_gene():
    gene_name = input("Introduza o nome do Gene (ex: ATM, HBB, NAT2): ").strip()
    
    conn = sqlite3.connect('disgenet.db')
    cursor = conn.cursor()
    
    # Query ajustada às tabelas reais do seu banco de dados
    query = """
    SELECT g.geneId, d.diseaseId, d.diseaseName, dc.diseaseClassName
    FROM geneAttributes g
    JOIN geneDiseaseNetwork f ON g.geneNID = f.geneNID
    JOIN diseaseAttributes d ON f.diseaseNID = d.diseaseNID
    LEFT JOIN disease2class d2c ON d.diseaseNID = d2c.diseaseNID
    LEFT JOIN diseaseClass dc ON d2c.diseaseClassNID = dc.diseaseClassNID
    WHERE g.geneName = ?
    """
    
    cursor.execute(query, (gene_name,))
    resultados = cursor.fetchall()
    
    if not resultados:
        print(f"\nNenhuma associação encontrada para o gene '{gene_name}'.")
    else:
        print(f"\n{'ID Gene':<10} | {'ID Doença':<12} | {'Nome da Doença':<45} | {'Tipo de Doença'}")
        print("-" * 105)
        for row in resultados:
            print(f"{row[0]:<10} | {row[1]:<12} | {row[2]:<45} | {str(row[3]).strip()}")
            
    conn.close()

if __name__ == "__main__":
    consultar_gene()
```
