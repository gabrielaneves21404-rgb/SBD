```python
import requests
```

```python


def gerar_lista_paises():
    # Otimizado com '?fields=...' para trazer apenas os dados necessários e evitar o Erro 400
    url = "https://restcountries.com/v3.1/all?fields=name,capital,currencies"
    response = requests.get(url)
    
    if response.status_code == 200:
        paises = response.json()
        
        # Cabeçalho da tabela com larguras fixas de caracteres
        print(f"{'Country':<30} {'Capital':<30} {'Currency'}")
        print("-" * 86)
        
        contador = 0
        for p in paises:
            # Extrair o nome comum
            nome_pais = p.get('name', {}).get('common', 'N/A')
            
            # Tratar a capital (que vem dentro de uma lista no JSON)
            lista_capitais = p.get('capital', ['N/A'])
            capital = lista_capitais[0] if lista_capitais else 'N/A'
            
            # Extrair dinamicamente o dicionário de moedas
            dict_moedas = p.get('currencies', {})
            moeda_formatada = "N/A"
            if dict_moedas:
                codigo_moeda = list(dict_moedas.keys())[0]
                nome_moeda = dict_moedas[codigo_moeda].get('name', '')
                moeda_formatada = f"{nome_moeda} ({codigo_moeda})"
                
            # Imprimir a linha formatada e alinhada à esquerda (<30)
            print(f"{nome_pais:<30} {capital:<30} {moeda_formatada}")
            contador += 1
            
        print("-" * 86)
        print(f"(Successfully retrieved {contador} countries)")
    else:
        print(f"Falha ao aceder à API. Código de estado: {response.status_code}")

if __name__ == "__main__":
    gerar_lista_paises()
```
